# 3. 비만율과 섭취하는 음식의 상관관계

가설: 비만율이 높은 나라가 비만율이 낮은 나라보다 영양소 섭취가 균일하지 못할 것이다.

전체 국가의 비만율의 평균을 구해 비만율이 높은 나라와 낮은 나라로 분류하여
비만율이 높은 나라와 낮은 나라의 평균 영양소 섭취량을 그래프로 그리기

```python
high_obesity = food[food.Obesity > food['Obesity'].mean()]#비만도의 평균값보다 큰 나라들 집합
low_obesity = food[food.Obesity <= food['Obesity'].mean()]#비만도의 평균값보다 작은 나라들 집합

animal_features = ['Animal fats', 'Aquatic Products, Other', 'Eggs', 'Fish, Seafood', 'Meat',
                   'Milk - Excluding Butter', 'Offals']
vegetal_features = ['Cereals - Excluding Beer', 'Fruits - Excluding Wine', 'Miscellaneous', 'Oilcrops', 'Pulses',
                    'Spices', 'Starchy Roots', 'Stimulants', 'Sugar & Sweeteners', 'Sugar Crops', 'Treenuts',
                    'Vegetable Oils', 'Vegetables']
```

```python
high_obesity = df[food_pure.Obesity > food_pure['Obesity'].mean()]#비만도의 평균값보다 큰 나라들 집합
low_obesity = df[food_pure.Obesity <= food_pure['Obesity'].mean()]#비만도의 평균값보다 작은 나라들 집합
```

# 3.1. 동물성 단백질, 식물성 단백질 섭취량과 비만의 상관 관계

가설: 비만율이 높은 국가의 동물성 단백질 섭취량이 비만율이 낮은 국가보다 높을 것이다.

비만율이 높은 국가의 동물성 단백질, 식물성 단백질 섭취량과 비만율이 낮은 국가의 동물성 단백질, 식물성 단백질 섭취량 비교

결과: 비만율이 높은 국가는 동물성 단백질과 식물성 단백질의 섭취량이 거의 반반이고, 비만율이 낮은 국가는 식물성 단백질 섭취량이 동물성 단백질 섭취량보다 높다.

```python
protein_high_obesity = protein[protein.Obesity > protein['Obesity'].mean()]
protein_low_obesity = protein[protein.Obesity <= protein['Obesity'].mean()]

p_high_animal = protein_high_obesity[animal_features].sum(axis=1).mean()
p_low_animal = protein_low_obesity[animal_features].sum(axis=1).mean()

p_high_vegetal = protein_high_obesity[vegetal_features].sum(axis=1).mean()
p_low_vegetal = protein_low_obesity[vegetal_features].sum(axis=1).mean()
```

```python
ratio = [p_high_vegetal, p_high_animal]
labels = ['Plant-based food', 'Animal-based food']

plt.pie(ratio, labels=labels, autopct='%.1f%%', counterclock=False, colors = ['#FD5D5D', '#C0EDA6'])
plt.title("Protein intake in high-obesity countries")

plt.savefig("high-obesity_protein.png", dpi=100)
```

```python
ratio = [p_low_vegetal, p_low_animal]
labels = ['Plant-based food', 'Animal-based food']

plt.pie(ratio, labels=labels, autopct='%.1f%%', counterclock=False, colors = ['#FD5D5D', '#C0EDA6'])
plt.title("Protein intake in low-obesity countries")

plt.savefig("low-obesity_protein.png", dpi=100)
```
